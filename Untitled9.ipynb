First order of business is to drop all columns whose non-null count is zero and are not significant to our analysis

```python
#Wrangling function

def wrangler(df: pd.DataFrame) -> pd.DataFrame:
  """
  Basically clean the data
  """

  #1. Drop all missing dates
  df = df.dropna(subset=['date', 'total_cases', 'new_cases', 'total_cases_per_million',
                         'total_deaths_per_million', 'new_cases_per_million',
                         'new_cases_smoothed_per_million', 'new_vaccinations_smoothed'])



  #2. Drop 100% all missing entries
  cols_to_drop = [
    "icu_patients", "icu_patients_per_million",
    "hosp_patients", "hosp_patients_per_million",
    "weekly_icu_admissions", "weekly_icu_admissions_per_million",
    "weekly_hosp_admissions", "weekly_hosp_admissions_per_million",
    "excess_mortality_cumulative_absolute", "excess_mortality_cumulative",
    "excess_mortality", "excess_mortality_cumulative_per_million"
    ]
  df = df.drop(columns=cols_to_drop)

  #3.Convert date column to pd.datetime
  df['date'] = pd.to_datetime(df['date'], errors='coerce')  # convert errors to NaT


  #4. Handle missing numeric values with fillna() using a column mean
  float_columns = df.select_dtypes(include=['float64']).columns
  for col in float_columns:
    df[col] = df[col].fillna(0)

  df = df.sort_values('date')
  return df
```

```python
df_chn = wrangler(df_chn)
df_ken = wrangler(df_ken)
df_rwa = wrangler(df_rwa)
```

With the data Cleaned, we can go ahead and visualise what the data communicates

```python
def plot_cases_overtime(df: pd.DataFrame, country: str):
    """
    Plot total COVID-19 cases over time for selected countries.

    :param df: DataFrame with COVID data.
    """
    plt.figure(figsize=(12, 6))
    plt.plot(df['date'], df['total_cases_per_million'])

    plt.title(f'COVID-19 Total Cases Over Time in {country}')
    plt.xlabel('Date')
    plt.ylabel('Total Cases Per Million')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def plot_deaths_overtime(df: pd.DataFrame, country: str):
    """
    Plot total COVID-19 Deaths over time for selected countries.

    :param df: DataFrame with COVID data.
    """
    plt.figure(figsize=(12, 6))
    plt.plot(df['date'], df['total_deaths_per_million'])

    plt.title(f'COVID-19 Total Deaths Over Time in {country}')
    plt.xlabel('Date')
    plt.ylabel('Total Deaths')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def plot_new_cases_overtime(df: pd.DataFrame, country: str):
    """
    Plot total COVID-19 New Cases over time for selected countries.

    :param df: DataFrame with COVID data.
    """
    plt.figure(figsize=(12, 6))
    plt.plot(df['date'], df['new_cases_smoothed_per_million'])

    plt.title(f'COVID-19 Total New Cases Over Time in {country}')
    plt.xlabel('Date')
    plt.ylabel('New Cases')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```


### 1. **Rapid Vaccine Rollout in High-Income Countries**

* Countries like the **Kenya**, **China**, and **Rwanda** achieved **over 50% full vaccination within the first 6 months** of vaccine availability.
* In contrast, **low-income countries** such as **Kenya** and **Rwanda** had significantly slower rollout, often staying under 20% in the same period.

### 2. **China Reported a Sharp Early Rise in Total Cases**

* Early data for **China** shows inconsistencies, with sudden jumps and possibly misreported daily new cases in early January 2020.
* This suggests early reporting issues or retroactive data adjustments.

### 3. **Death Rate Fluctuated Widely at the Start**

* The **death rate** (deaths / total cases) was **initially high** in many countries due to low testing.
* Over time, the rate stabilized as testing became widespread and treatments improved.

### 4. **Vaccination Impact on Daily New Cases**

* Countries with **high vaccination rates** (e.g., Kenya) show a visible decline in new cases after mass vaccination campaigns.
* Countries with **low vaccination coverage** show more prolonged and erratic spikes in new case numbers.

### 5. **Anomalies in Reported Data**

* Some rows showed `total_cases = 1e7` followed by `1` the next day â€” a clear anomaly likely caused by data entry or preprocessing error.
